```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("08-seismic_activity_svm.csv")
```

```python
df.info()
```

```python

# Veri boyutu ve NaN kontrolü
print("Satır sayısı:", len(df))
print("NaN değerler:\n", df[['underground_wave_energy','vibration_axis_variation','seismic_event_detected']].isnull().sum())
print("Sonsuz değerler:\n", df[['underground_wave_energy','vibration_axis_variation']].apply(lambda c: (~c.between(-1e15, 1e15)).sum()))
print(df['seismic_event_detected'].value_counts())
```

```python
sns.scatterplot(
    data=df,
    x='underground_wave_energy',
    y='vibration_axis_variation',
    hue='seismic_event_detected',
    alpha=0.5,
    s=20
)
plt.title("Seismic Activity")
plt.tight_layout()
```
